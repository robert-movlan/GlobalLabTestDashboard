```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np

# Load CSV
df = pd.read_csv("shipment_table.csv")


# Preview
print("Initial Data:")
display(df.head())

# Step 1: Normalize priority to numerical weight
priority_weights = {"Low": 1, "Medium": 2, "High": 3, "Critical": 4}
df["priority_score"] = df["priority_level"].map(priority_weights)

# Step 2: Create result flags
df["is_positive"] = df["result_outcome"].apply(lambda x: 1 if x == "Positive" else 0)
df["is_inconclusive"] = df["result_outcome"].apply(lambda x: 1 if x == "Inconclusive" else 0)

# Step 3: Flag turnaround thresholds
df["fast_turnaround"] = df["turnaround_hours"].apply(lambda x: 1 if x <= 12 else 0)
df["slow_turnaround"] = df["turnaround_hours"].apply(lambda x: 1 if x >= 48 else 0)

# Step 4: Add time bucket
df["test_timestamp"] = pd.to_datetime(df["test_timestamp"])
df["day_of_week"] = df["test_timestamp"].dt.day_name()
df["hour_bucket"] = pd.cut(df["test_timestamp"].dt.hour,
                           bins=[0, 6, 12, 18, 24],
                           labels=["Night", "Morning", "Afternoon", "Evening"],
                           right=False)

# Final Review
print("Transformed Data:")
display(df.head())

# Optional: Save back to JSON
df.to_json("shipment_transformed.json", orient="records", indent=2, date_format='iso')
print("✅ shipment_transformed.json saved!")
```
